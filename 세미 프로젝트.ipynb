#### *'~시'의 경우 구한 요소들 : 시청소재지, 시장, 인구, 인구밀도, 면적, 시목, 시화, 시조, 마스코트*

```python
import requests
from bs4 import BeautifulSoup
address = 'https://ko.wikipedia.org/wiki/'
url = address + '서울특별시'
res = requests.get(url)
html = res.content
soup = BeautifulSoup(html, 'html.parser')
infobox = soup.select('table.infobox > tbody > tr')

seoul_list = []

for info in infobox:
    seoul_list.append(info.text.replace('\n', ''))
```

```python
result_set = []
admin_list = ['서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', \
              '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원도_(남)',\
             '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']

for i in range(len(admin_list)):
    
    address = 'https://ko.wikipedia.org/wiki/'
    url = address + admin_list[i]
    res = requests.get(url)
    html = res.content
    soup = BeautifulSoup(html, 'html.parser')
    infobox = soup.select('table.infobox > tbody > tr')

    data_set = []
    for data in infobox:
        source = data.text
        source =  source.split('\n')
        source = [v for v in source if v]
        data_set.append(source)
    
    data_result = []

    if admin_list[i][-1] == '시':
        for data in data_set:
            if '시청소재지 ' in data:
                data_result.append(data)
            elif '시장 ' in data:
                data_result.append(data)
            elif '인구 ' in data:
                data_result.append(data)
            elif '인구밀도 ' in data:
                data_result.append(data)
            elif '면적 ' in data:
                data_result.append(data)
            elif '시목 ' in data:
                data_result.append(data)
            elif '시화 ' in data:
                data_result.append(data)
            elif '시조 ' in data:
                data_result.append(data)
            elif '마스코트 ' in data:
                data_result.append(data)
                
        result_set.append(data_result)
    else:
        for data in data_set:
            if '소재지' in data:
                data_result.append(data)
            elif '도지사' in data: 
                data_result.append(data)
            elif '인구' in data:
                data_result.append(data)
            elif '인구밀도' in data:
                data_result.append(data)
            elif '면적' in data:
                data_result.append(data)
            elif '도목' in data:
                data_result.append(data)
            elif '도화' in data:
                data_result.append(data)
            elif '도조' in data:
                data_result.append(data)
        
        result_set.append(data_result)

result_set.append(data_result)
```

## 대한민국의 행정 구역 페이지에서 각 행정구역마다 a태그에 있는 href 주소를 리스트화해서 담겠다

```python
url = 'https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%ED%96%89%EC%A0%95_%EA%B5%AC%EC%97%AD'
res = requests.get(url)
html = res.content
soup = BeautifulSoup(html, 'html.parser')
loc_address = soup.select('#mw-content-text > div.mw-parser-output > ul > li  a')
address_set = []   # 주소
name_set = []  # 지역명

si = []          # 시, 군 , 구로 나눠서 하자 어차피 나중에 시목, 구목 다 다르니까.
si_name = []
gun = []
gun_name = []
gu = []
gu_name = []
etc = []

i = 0 

for address in loc_address:
    print(address.text, " : ", address['href'])  # 지역과 해당 주소를 프린트

    if address.text[-1] == '시':
        si.append(address['href']) # 주소 넣고
        si_name.append(address.text) # 지역명 넣고
    elif address.text[-1] == '군':
        gun.append(address['href'])
        gun_name.append(address.text)
    elif address.text[-1] == '구':
        gu.append(address['href'])
        gu_name.append(address.text)
    else:
        etc.append(address.text) # 혹시 모르니 나머지 것들 담아둠
        etc.append(address['href'])
        
    i += 1   # 보니까 지역만 가져오는게 아니라 밑에 잡다한게 딸려옴. 제주도 서귀포시를 끝으로 안받고 싶으므로 269까지만 반복해서 가져온다. 
    if i == 269:
        break
```

# 이제 구한 주소별로 request해서 원하는 정보를 크롤링한다

## 시
행정구역, 시청소재지, 시장, 세대, 인구, 총인구, 인구밀도, 소재지, 면적, 시목, 시화, 시조, 마스코트

```python
address = 'http://ko.wikipedia.org'
si_dict_set = [] # 딕셔너리들 담을 곳

for i in range(len(si)):
    condition = si[i]
    url = address + condition
    res = requests.get(url)
    html = res.content
    soup = BeautifulSoup(html, 'html.parser')
    infobox = soup.select('table.infobox > tbody > tr')
    
    
    si_dict = {
    '행정구역' : '', '시청소재지' : '', '시장' :'', '세대':'', '인구':'', '총인구' :'', '인구밀도' : '', '소재지' : '', '면적' : '', '시목' : '', '시화' : '', '시조' : '', '마스코트' : ''
}
    si_set = []
    for data in infobox:
        source = data.text
        source =  source.split('\n')
        source = [v for v in source if v]
        if len(source) ==2:
            source[0] = source[0].replace(' ', '')
            si_set.append(source)


    for data in si_set:
        if data[0] in si_dict.keys():
            si_dict[data[0]] = data[1]
    
    si_dict_set.append(si_dict)
```
